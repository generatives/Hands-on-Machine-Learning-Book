```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1)
```

```python
X, y = mnist["data"], mnist["target"]
```

```python
import numpy as np

y = y.astype(np.uint)
```

```python
X_train, y_train, X_test, y_test = X[:60000], y[:60000], X[60000:], y[60000:]
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, verbose=0)
```

```python
%time rf_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, rf_clf.predict(X_test))
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
```

```python
pca.fit(X_train)
```

```python
X_train_pca = pca.transform(X_train)
```

```python
X_train_pca.shape
```

```python
X_train.shape
```

```python
rf_clf_pca = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1, verbose=0)
```

```python
%time rf_clf_pca.fit(X_train_pca, y_train)
```

```python
X_test_pca = pca.transform(X_test)
accuracy_score(y_test, rf_clf_pca.predict(X_test_pca))
```

```python
from sklearn.manifold import TSNE

tSNE = TSNE(n_components=2, verbose=2)
```
